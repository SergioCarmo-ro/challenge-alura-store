```python
# Calcular a média das avaliações
avaliacao_compra = {'Assistente virtual.csv' : 4, 'Mesa de jantar.csv' : 1, 'Jogo de tabuleiro.csv' : 1, 'Micro-ondas.csv' : 4, 'Cadeira de escritório.csv' : 5}
```

```python
# somatorio das medias
somatorio = sum(avaliacao_compra.values())
somatorio
```

```python
# media do somatorio das medias
media = somatorio / len(avaliacao_compra)
media
```

```python
# Grafico de barras das medias e mostra da media das medias

import matplotlib.pyplot as plt

# Dados de avaliação
avaliacao_compra = {'Assistente virtual.csv': 4, 'Mesa de jantar.csv': 1, 'Jogo de tabuleiro.csv': 1, 'Micro-ondas.csv': 4, 'Cadeira de escritório.csv': 5}

# Calcula a média das avaliações
medias = list(avaliacao_compra.values())
media_das_medias = sum(medias) / len(medias)

# Criando o gráfico de barras
plt.figure(figsize=(6, 4))  # Define o tamanho da figura
plt.bar(avaliacao_compra.keys(), medias, color='skyblue')
plt.axhline(y=media_das_medias, color='r', linestyle='--', label=f'Média das Médias: {media_das_medias:.2f}')

plt.xlabel('Produtos')
plt.ylabel('Avaliação')
plt.title('Média de Avaliação dos Produtos')
plt.xticks(rotation=45, ha='right')  # Rotaciona os rótulos do eixo x para melhor legibilidade
plt.legend()
plt.tight_layout() # Ajusta o layout para evitar sobreposição de elementos
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
import pandas as pd
import matplotlib.pyplot as plt
loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

```python

#Concatenando os datasets
df = pd.concat([loja, loja2, loja3, loja4])
#Contagem de produtos vendidos
produtos_vendidos = df['Produto'].value_counts()

# Produto mais vendido
mais_vendido = produtos_vendidos.idxmax()
quantidade_mais_vendido = produtos_vendidos.max()

# Produto menos vendido
menos_vendido = produtos_vendidos.idxmin()
quantidade_menos_vendido = produtos_vendidos.min()

print(f"Produto mais vendido: {mais_vendido} ({quantidade_mais_vendido} unidades)")
print(f"Produto menos vendido: {menos_vendido} ({quantidade_menos_vendido} unidade{'s' if quantidade_menos_vendido > 1 else ''})")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing sales data
# Replace 'df' with the actual name of your DataFrame if different
df = pd.concat([loja, loja2, loja3, loja4])


# Contagem de produtos vendidos
produtos_vendidos = df['Produto'].value_counts()

# Top 10 produtos mais vendidos
_mais_vendidos = produtos_vendidos.head(10).values # Get the values for the pie chart
mais_vendidos = produtos_vendidos.head(10) # Get the Series for labels

# Top 10 produtos menos vendidos
_menos_vendidos = produtos_vendidos.tail(10).values # Get the values for the pie chart
menos_vendidos = produtos_vendidos.tail(10) # Get the Series for labels


# Gráfico de pizza - Mais vendidos
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.pie(_mais_vendidos, labels=mais_vendidos.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Produtos Mais Vendidos')
plt.axis('equal')

# Gráfico de pizza - Menos vendidos
plt.subplot(1, 2, 2)
plt.pie(_menos_vendidos, labels=menos_vendidos.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Produtos Menos Vendidos')
plt.axis('equal')

plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


```python
import pandas as pd
loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

```python
# Frete Médio por Loja

# Calculo do custo médio de envio para cada loja
média_frete_loja1 = loja['Frete'].mean()
média_frete_loja2 = loja2['Frete'].mean()
média_frete_loja3 = loja3['Frete'].mean()
média_frete_loja4 = loja4['Frete'].mean()

# Imprimindo os Resultados
print(f"Custo médio de envio para Loja 1: R${média_frete_loja1:.2f}")
print(f"Custo médio de envio para Loja 2: R${média_frete_loja2:.2f}")
print(f"Custo médio de envio para Loja 3: R${média_frete_loja3:.2f}")
print(f"Custo médio de envio para Loja 4: R${média_frete_loja4:.2f}")

# Criando um gráfico de barras para visualizar os custos médios de envio
import matplotlib.pyplot as plt

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
custos_médios_de_envio = [média_frete_loja1, média_frete_loja2, média_frete_loja3, média_frete_loja4]

plt.figure(figsize=(6, 4))
plt.bar(lojas, custos_médios_de_envio, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.xlabel("Lojas")
plt.ylabel("Custo médio de envio (R$)")
plt.title("Custo médio de envio por loja")
plt.show()
```

```python
# 3 gráficos:  um de pizza sobre  Faturamento Total de Cada Loja,
# um de Barras vertical sobre  Produtos mais e menos vendidos
# um de barras horizontal sobre o custo medio do Frete.

import pandas as pd
import matplotlib.pyplot as plt


# Criando coluna 'Faturamento' para loja2, loja3 e loja4
loja['Faturamento'] = loja['Preço'] + loja['Frete']
loja2['Faturamento'] = loja2['Preço'] + loja2['Frete']
loja3['Faturamento'] = loja3['Preço'] + loja3['Frete']
loja4['Faturamento'] = loja4['Preço'] + loja4['Frete']

# Dados para o gráfico de pizza (Faturamento Total de Cada Loja)
faturamento_lojas = {
    'Loja 1': loja['Faturamento'].sum(),
    'Loja 2': loja2['Faturamento'].sum(),
    'Loja 3': loja3['Faturamento'].sum(),
    'Loja 4': loja4['Faturamento'].sum()
}

# Criando o gráfico de pizza
plt.figure(figsize=(7, 7))
plt.pie(faturamento_lojas.values(), labels=faturamento_lojas.keys(), autopct='%1.1f%%', startangle=90)
plt.title('Faturamento Total de Cada Loja')
plt.axis('equal')  # proporção de aspecto igual garante que a pizza seja como um círculo
plt.show()


# Dados para o gráfico de barras vertical (Produtos mais e menos vendidos)
# Você já possui 'produtos_vendidos' com a contagem dos produtos.
# Vamos usar os 5 mais vendidos e os 5 menos vendidos para melhor visualização.

top_5_produtos = produtos_vendidos.head(5)
bottom_5_produtos = produtos_vendidos.tail(5)

plt.figure(figsize=(9, 4))

plt.subplot(1, 2, 1)
plt.bar(top_5_produtos.index, top_5_produtos.values)
plt.title('Top 5 Produtos Mais Vendidos')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
plt.bar(bottom_5_produtos.index, bottom_5_produtos.values)
plt.title('Top 5 Produtos Menos Vendidos')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()


# Dados para o gráfico de barras horizontal (Custo médio do frete)
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
custos_medios_frete = [média_frete_loja1, média_frete_loja2, média_frete_loja3, média_frete_loja4]

plt.figure(figsize=(5, 3))
plt.barh(lojas, custos_medios_frete, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.xlabel('Custo Médio do Frete (R$)')
plt.ylabel('Lojas')
plt.title('Custo Médio do Frete por Loja')
plt.show()
```
